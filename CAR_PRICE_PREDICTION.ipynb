```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('car_prices.csv')
df.head()

df = df.drop('Year', axis=1)
df

df.columns = ['car_maker','car_model','engine_size','hp','torque','seconds_to_60mph','price']
print(df.head())

df = df.head(50)
df
```

```python
df['car_model'].value_counts()
```

```python
plt.scatter(x=df.hp,y=df['price'])
plt.show()
```

```python
plt.scatter(x=df.torque,y=df['price'])
plt.show()
```

```python
plt.bar(df.car_model,df.price)
plt.show()
```

```python
dummies = pd.get_dummies(df.car_model)
dummies
```

```python
merged_df = pd.concat([df,dummies],axis=1)
merged_df
```

```python
	4C	4C Spider	A110	Atom	DB11	DBS Superleggera	Giulia Quadrifoglio	NSX	Vantage
```

```python
final_df = merged_df.drop(columns=['car_model','engine_size','seconds_to_60mph','DB11','NSX','4C','4C Spider','A110','DBS Superleggera','Giulia Quadrifoglio'])
final_df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=final_df[['hp','torque','Atom','Vantage']]
y=final_df['price']
```

```python
model=LinearRegression()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

y = pd.to_numeric(y.str.replace(',', ''), errors='coerce')

model.fit(x, y)
```

```python
model.score(x,y)
```
